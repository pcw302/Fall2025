4) The data shown is actually from several years. Let's turn this into an animated plot that will show each year separately. Add the argument: `animation_frame = 'Year'` to px.scatter. <br>

```python
max_x,max_y = tab_2000['Maternity'].max(), tab_2000['Paternity'].max()
max_x *= 1.05
max_y *= 1.05
min_x, min_y = -5,-5

fig = px.scatter(tab_2000,
                 x='Maternity',
                 y='Paternity',
                 size='Protected',
                 animation_frame='Year',
                 color='Continent',
                 hover_name='Country',
                 color_discrete_sequence=px.colors.qualitative.Set2,
                 labels={
                     'Maternity':'Maternity Leave (weeks) ',
                     'Paternity':'Paternity Leave (weeks) ',
                     'Protected':'Protected leave in weeks '},
                 hover_data={'Continent': False, 'Maternity': ':.1f', 'Paternity': ':.1f', 'Protected': ':.1f'},)

fig.update_layout(
    title='Maternity vs Paternity Leave (2000-2020)',
    xaxis=dict(range=[min_x, max_x]),
    yaxis=dict(range=[min_y, max_y]),
    template='plotly_white',
)

fig.show()
```
